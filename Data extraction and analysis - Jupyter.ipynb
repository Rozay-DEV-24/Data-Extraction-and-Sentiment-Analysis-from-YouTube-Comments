
> 2.b  Provide the following statistics for query string “avatar movie” of top 50 videos sorted by relevance in the US region ( 7 marks )

> Output expected: video ID, title, no of views, no of likes,no of comments exported to CSV file






Reference link: https://developers.google.com/youtube/v3/docs/videos/list

```python
from googleapiclient.discovery import build
import csv

api_key = 'AIzaSyBVSPZJApn4ufVJ4xgSesw2F2iKpwVXV_8'

youtube = build('youtube', 'v3', developerKey=api_key)

search_request = youtube.search().list(
    part='id,snippet',  
    q='avatar movie',
    type='video',
    regionCode='US',
    maxResults=50
)

search_response = search_request.execute()

video_info = []  

for item in search_response.get('items', []):
    video_id = item['id']['videoId']
    title = item['snippet']['title']
    video_info.append((video_id, title))

stats_request = youtube.videos().list(
    part='statistics',
    id=','.join(video_id for video_id, _ in video_info)
)

stats_response = stats_request.execute()


for item in stats_response['items']:
    video_id = item['id']
    stats = item['statistics']
    title = next(title for vid, title in video_info if vid == video_id)
    views = stats.get('viewCount', 0)
    likes = stats.get('likeCount', 0)
    comments = stats.get('commentCount', 0)

    print(f"Video ID: {video_id}")
    print(f"Title: {title}")
    print(f"Views: {views}")
    print(f"Likes: {likes}")
    print(f"Comments: {comments}")
    print("_" * 100)


with open('Question2b.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Video ID', 'Title', 'Views', 'Likes', 'Comments']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for item in stats_response['items']:
        video_id = item['id']
        stats = item['statistics']
        title = next(title for vid, title in video_info if vid == video_id)
        views = stats.get('viewCount', 0)
        likes = stats.get('likeCount', 0)
        comments = stats.get('commentCount', 0)

        writer.writerow({
            'Video ID': video_id,
            'Title': title,
            'Views': views,
            'Likes': likes,
            'Comments': comments
        })
print("Output data exported as Question2b.json")
```

```python
from googleapiclient.discovery import build
import csv
import json

api_key = 'AIzaSyBVSPZJApn4ufVJ4xgSesw2F2iKpwVXV_8'

youtube = build('youtube', 'v3', developerKey=api_key)

top_video_data = []
with open('top_commented_videos.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        top_video_data.append(row)

output_data = []

for video_info in top_video_data:
    video_id = video_info['Video ID']
    
    comment_request = youtube.commentThreads().list(
        part='snippet',
        videoId=video_id,
        maxResults=10
    )
    
    comment_response = comment_request.execute()
    
    comments = [
        {
            "id": item['id'],
            "kind": item['kind'],
            "snippet": {
                "textDisplay": item['snippet']['topLevelComment']['snippet']['textDisplay']
            }
        }
        for item in comment_response.get('items', [])
    ]
    
    output_data.append({
        "video_id": video_id,
        "comments": comments
    })

# Save the output_data to a JSON file
with open('Question3c.json', 'w') as jsonfile:
    json.dump(output_data, jsonfile, indent=4)

print("Output data saved to Question3c.json")

```

>3.d Write a function to get  the likes vs views ratio of the top 10 videos obtained in 3.a with the highest comments (3 marks)




```python
from googleapiclient.discovery import build
import csv

api_key = 'AIzaSyBVSPZJApn4ufVJ4xgSesw2F2iKpwVXV_8'

def get_likes_views_ratio(api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)

    search_request = youtube.search().list(
        part='id,snippet',  
        q='avatar movie',
        type='video',
        regionCode='US',
        maxResults=10
    )

    search_response = search_request.execute()

    video_info = []  

    for item in search_response.get('items', []):
        video_id = item['id']['videoId']
        title = item['snippet']['title']
        video_info.append((video_id, title))

    stats_request = youtube.videos().list(
        part='statistics',
        id=','.join(video_id for video_id, _ in video_info)
    )

    stats_response = stats_request.execute()

    video_data = []

    for item in stats_response['items']:
        video_id = item['id']
        stats = item['statistics']
        title = next(title for vid, title in video_info if vid == video_id)
        views = int(stats.get('viewCount', 0))
        likes = int(stats.get('likeCount', 0))
        comments = int(stats.get('commentCount', 0))
        likes_to_views_ratio = likes / views if views > 0 else 0

        video_data.append({
            "video_id": video_id,
            "title": title,
            "views": views,
            "likes": likes,
            "comments": comments,
            "likes_to_views_ratio": likes_to_views_ratio
        })

    sorted_videos = sorted(video_data, key=lambda x: x['comments'], reverse=True)

    return sorted_videos

top_videos_with_ratio = get_likes_views_ratio(api_key)

for video in top_videos_with_ratio:
    print(f"Video ID: {video['video_id']}")
    print(f"Title: {video['title']}")
    print(f"Views: {video['views']}")
    print(f"Likes: {video['likes']}")
    print(f"Comments: {video['comments']}")
    print(f"Likes to Views Ratio: {video['likes_to_views_ratio']:.4f}")
    print("_" * 100)


```
